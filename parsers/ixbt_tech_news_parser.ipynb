```python
from datetime import datetime, timedelta
```

```python
date_str = "2025/09/15"
cur_date = datetime.strptime(date_str, "%Y/%m/%d").date()

def calc_date(date: datetime):
    date = date - timedelta(days=1)
    return date, date.strftime("%Y/%m/%d")
```

```python
import requests
import time
from bs4 import BeautifulSoup
from random import uniform
import os
import re
```

```python
site_url = "https://ixbt.com"
base_url = site_url+"/news/"

DATA_FOLDER = 'parsed_tech_news'

connect_timeout = 100
read_timeout = 100
```

```python
parsed_dir = os.path.join(os.getcwd(), DATA_FOLDER)
os.makedirs(parsed_dir, exist_ok=True)
```

```python
def get_soup(url: str):
    response = requests.get(url, verify=False, timeout=(connect_timeout, read_timeout))
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup
```

```python
counter = 1
while True:
    #  
    cur_date, date_str = calc_date(cur_date)

    #    
    url = base_url + date_str

    date_str = date_str.replace("/", "-")

    hub_soup = get_soup(url)

    link_data = hub_soup.find('div', class_ = re.compile("b-block block__newslistdefault b-lined-title"))
    link_data = link_data.ul
    links = link_data.find_all('a', href=True)
    all_links = [link['href'] for link in links]

    #     
    for link in all_links:
        file_path = os.path.join(parsed_dir, str(date_str) + "_" + str(counter) + ".html")

        article_soup = get_soup(site_url+link)

        #    
        title_data = article_soup.find('div', class_=re.compile('b-article__header'))
        title = title_data.find('h1')

        #    
        text_data = article_soup.find('div', class_=re.compile('b-article__content'))
        text_list = text_data.find_all('p') 

        #   
        f = open(file_path, 'w', encoding="utf-8")

        f.write(str(title))
        for elem in text_list:
            f.write(str(elem))

        wait_time = uniform(0, 3)
        time.sleep(wait_time)

        f.close()

        print(f"Webpage {counter} parsed")

        counter += 1

    print(f"Date {date_str} cleared")

    counter = 1
```
